```python
sec_9 = ['er_avail_adrenalin',
'er_avail_atropine',
'er_avail_dopamine',
'er_avail_cephalosporins_betalactams',
'er_avail_aminoglycosides_quinolone',
'er_avail_vancomycin_clindamycin',
'er_avail_lidocaine',
'er_avail_minor_opioids',
'er_avail_major_opioids',
'er_avail_iv_fluids',
'er_avail_diazepam_dph',
'er_avail_heparin',
'er_avail_steroids',
'er_avail_insulin',
'er_avail_asthma',
'er_avail_blood_pressure',
'er_avail_defibrillator',
'er_avail_ott_intubation',
'er_avail_catheter',
'er_avail_oxygen_suction']


sec_9_df = df[sec_9].copy()

# Drop Never existed
sec_9_df[sec_9_df == 'Nunca ha existido'] = np.nan
sec_9_df = sec_9_df.dropna()

# Change Categorical Values to Ordinal and Numerical Values
sec_9_df[sec_9_df == 'Todos los d√≠as'] = 3
sec_9_df[sec_9_df == 'Entre 3 y 5 d√≠as'] = 2
sec_9_df[sec_9_df == 'Entre 1 y 2 d√≠as'] = 1
sec_9_df[sec_9_df == 'No hubo'] = 0

# Create Sum of each row
sec_9_df['sum'] = sec_9_df.sum(axis = 1)

# Create scores of each week
sec_9_df['ScoreWeek'] = 100 - ((sec_9_df['sum']/(3*len(sec_9))*100))

sec_9_df
```
